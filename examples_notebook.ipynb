# Some examples using the Hydro Class

First load the Hydro Class from the package.  

To check out possible functions, do self. and push tab.  
To check out parameters do shift+tab within () of function to open docstring.

```python
from wsbd import Hydro
%matplotlib inline
```

Create class instance to start using the API-client.

```python
h = Hydro()
```

Return the themes from HydroNET API.

```python
h.return_themes()
```

Return the parameters for any themeid, see above.

```python
h.return_parameters(9)
```

Download all geolocations for theme to json.

```python
h.download_geolocations(2, 'geolocations.json')
```

Return geolocations, for theme, possible filters can be added to filter on code and name.

```python
h.return_geolocations(2)
```

```python
h.return_geolocations(2, codefilter='TDB') # to return all discharge stations
```

```python
h.return_geolocations(2, namefilter='Brooymans') # to search for names
```

```python
h.return_geolocations(2, codefilter='TDB', namefilter='Brooymans') # can also be combined
```

Return possible timeseries for given theme

```python
h.return_timeseries(2)
```

Return timeseries, other functions can be used to built request. Default return series, from 1 January this year until now.

```python
brooymans = h.return_geolocations(2, codefilter='TDB', namefilter='Brooymans')
parameters = h.return_parameters(2)
h.return_data(theme=2,parameter=parameters[parameters.Description.str.contains('Debiet')]['ParameterID'].iloc[0],\
              location=brooymans.LocationID.iloc[0]).head()
```

values=False, also return metadata

```python
brooymans = h.return_geolocations(2, codefilter='TDB', namefilter='Brooymans')
parameters = h.return_parameters(2)
h.return_data(theme=2,parameter=parameters[parameters.Description.str.contains('Debiet')]['ParameterID'].iloc[0],\
              location=brooymans.LocationID.iloc[0],values=False).head()
```

startdate and enddate can be used to specify period default is english string data, for example '20180101'

```python
h.return_data(theme=2,parameter=1318,location=52300,startdate='20180101',enddate='20190101').head()
```

self.return_data() returns a dataframe therefor .plot() or other functions can be used directly on the result, of values=True.

```python
h.return_data(theme=2,parameter=1318,location=52300,startdate='20180101',enddate='20190101').plot()
```

# Basic usage kiwis_pie
Python library for querying WISKI via KiWIS (KISTERS Web Interoperability Solution).

```python
from datetime import date
from kiwis_pie import KIWIS

k = KIWIS('http://10.10.3.126:8080/KiWIS/KiWIS') # set kiwis server
```

```python
k.get_station_list(stationparameter_name='Debiet') # all stations where discharge is measured
```

```python
k.get_timeseries_list(stationparameter_name='Debiet', ts_name='Dag.Gem') # all discharge stations, with series Dag.Gem
```

```python
ts_id = k.get_timeseries_list(stationparameter_name='Debiet', ts_name='Dag.Gem')['ts_id'].values[0] # ts_id of first discharge station with series Dag.Gem
k.get_timeseries_values(ts_id=ts_id, to=date(2017,1,1), **{'from': date(2016,1,1)}) # get series from 1-1-2016 until 1-1-2017, you could also use period='p1y'
```

```python
series = k.get_timeseries_values(ts_id=ts_id, to=date(2017,1,1), **{'from': date(2016,1,1)})
series.index = series.index.tz_localize(None) # to get rid of the timezone
series
```

# Basic usage of hkvsobekpy & geopandas
Loading and visualizing data from SOBEK.

```python
import hkvsobekpy as hkv

his_file = 'test_data\\hkvsobekpy\\CALCPNT.HIS' # binary his-file
calcpnt = hkv.read_his.ReadMetadata(his_file).DataFrame() # read the file, convert to dataframe
calcpnt
```

```python
waterlvl = calcpnt.T.loc['Waterlevel  (m AD)  '] # waterlevels and transpose, such that index is equal to nodes
waterlvl
```

```python
import geopandas as gpd

network_file = 'test_data\\hkvsobekpy\\NETWORK_n.shp'
gdf = gpd.read_file(network_file) # read the node network file
gdf.crs = {'init': 'epsg:28992'}
gdf
```

```python
import pandas as pd

waterlvl = gdf.merge(waterlvl, left_on='ID', right_index=True, how='right') # merge waterlevels on network nodes
columns = [str(column) for column in waterlvl.columns] # from timestamp columns to string
waterlvl.columns = columns
waterlvl
```

```python
import matplotlib.pyplot as plt

ax = waterlvl.plot(column='2013-02-20 00:00:00', figsize=(16,12), legend=True, cmap='jet', legend_kwds={'shrink': 0.5}) # plot waterlevel on last timestep
for x, y, label in zip(waterlvl.geometry.x, waterlvl.geometry.y, waterlvl['2013-02-20 00:00:00']): # annotate the waterlevel of last timestep
    ax.annotate(round(label,2), xy=(x, y), xytext=(3, 3), textcoords="offset points")
plt.title('waterstand (m NAP)', size=16)
plt.axis('equal')
plt.axis('off');
```
